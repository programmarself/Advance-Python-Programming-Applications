```python
import numpy as np

# Define a coefficient matrix (A) and a right-hand side vector (b)
A = np.array([[3, 1], [1, 2]])  # 2x2 matrix
b = np.array([9, 8])            # 2-element vector

# Solve the linear system: A * x = b
x = np.linalg.solve(A, b)       # x holds the solution of the system

print("Solution of A*x=b:", x)   # Output the solution
```

```python
import numpy as np

# Define a square matrix
matrix = np.array([[4, 7], [2, 6]])

# Compute its inverse using np.linalg.inv
matrix_inv = np.linalg.inv(matrix)  # Get the inverse of the matrix

print("Matrix Inverse:\n", matrix_inv)  # Display the inverse
```

```python
import numpy as np

# Define a square matrix
matrix = np.array([[4, 7], [2, 6]])

# Calculate the determinant using np.linalg.det
det = np.linalg.det(matrix)  # Compute the determinant

print("Determinant of the matrix:", det)  # Output the determinant
```

```python
import numpy as np

# Define a vector or matrix
vector = np.array([3, 4])

# Compute the Euclidean norm (L2 norm) of the vector using np.linalg.norm
norm_value = np.linalg.norm(vector)

print("Euclidean Norm:", norm_value)  # Expected output: 5, since sqrt(3^2 + 4^2) = 5
```

```python
import numpy as np

# Define a square matrix
matrix = np.array([[12, -51, 4],
                   [6, 167, -68],
                   [-4, 24, -41]])

# Compute QR decomposition of the matrix
Q, R = np.linalg.qr(matrix)  # Decomposes matrix into orthogonal matrix Q and upper triangular matrix R

print("Matrix Q (orthogonal):\n", Q)
print("Matrix R (upper triangular):\n", R)
```

```python
import numpy as np

# Define a square matrix
matrix = np.array([[5, 4], [1, 2]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(matrix)  # Perform eigen decomposition

print("Eigenvalues:", eigenvalues)          # Output the eigenvalues
print("Eigenvectors:\n", eigenvectors)        # Output the eigenvectors (columns)
```
